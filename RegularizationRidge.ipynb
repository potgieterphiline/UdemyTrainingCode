```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/Advertising.csv")
```

```python
X = df.drop(['sales'], axis=1)
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
poly_conv = PolynomialFeatures(degree=3, include_bias=False)
```

```python
poly_feat = poly_conv.fit_transform(X)
```

```python
poly_feat.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_feat, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train[0]
```

```python
poly_feat
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train,y_train)
```

```python
test_pred = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, test_pred)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
```

```python
RMSE
```

```python
from sklearn.linear_model import RidgeCV
```

```python
ridge_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0),scoring = 'neg_mean_absolute_error')
```

```python
ridge_cv_model.fit(X_train, y_train)
```

```python
ridge_cv_model.alpha_
```

```python
test_pred_2 = ridge_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_pred_2)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_pred_2))
```

```python
RMSE
```

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv_model = LassoCV(eps=0.1, n_alphas=100, cv=5)
```

```python
lasso_cv_model.fit(X_train, y_train)
```

```python
lasso_cv_model.alpha_
```

```python
test_predictions = lasso_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
```
