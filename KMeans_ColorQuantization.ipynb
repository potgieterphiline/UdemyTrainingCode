```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.image as mpimg
```

```python
image_array = mpimg.imread('/content/sample_data/palm_trees.jpg')
```

```python
image_array.shape
```

```python
plt.imshow(image_array)
```

```python
(h,w,c) = image_array.shape
```

```python
image_array_2d = image_array.reshape(h*w,c)
```

```python
image_array_2d
```

```python
image_array_2d.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6)
```

```python
labels = model.fit_predict(image_array_2d)
```

```python
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
quant_image = np.reshape(rgb_codes[labels],(h,w,c))
```

```python
plt.imshow(quant_image)
```
