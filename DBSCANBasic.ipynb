```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blobs = pd.read_csv('/content/sample_data/cluster_blobs.csv')
moons = pd.read_csv('/content/sample_data/cluster_moons.csv')
circles = pd.read_csv('/content/sample_data/cluster_circles.csv')
```

```python
blobs.head()
```

```python
sns.scatterplot(data=blobs,x='X1',y='X2')
```

```python
sns.scatterplot(data=moons,x='X1',y='X2')
```

```python
sns.scatterplot(data=circles,x='X1',y='X2')
```

```python
def display_cats(model,data):
  labels = model.fit_predict(data)
  sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
display_cats(model,blobs)
```

```python
model = KMeans(n_clusters=2)
display_cats(model,moons)
```

```python
model = KMeans(n_clusters=2)
display_cats(model,circles)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN()
display_cats(model,blobs)
```

```python
model = DBSCAN(eps=0.15)
display_cats(model,moons)
```

```python
model = DBSCAN(eps=0.15)
display_cats(model,circles)
```
