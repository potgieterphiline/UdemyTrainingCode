```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/sample_data/cluster_mpg.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_w_dummies = pd.get_dummies(df.drop('name', axis=1))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_w_dummies)
```

```python
scaled_df = pd.DataFrame(df_scaled, columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(scaled_df)
```

```python
plt.figure(figsize=(2,8))
sns.clustermap(scaled_df,col_cluster = False);
```

```python
sns.heatmap(scaled_df.corr())
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=6)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
sns.scatterplot(data=df,x='mpg',y='horsepower',hue=cluster_labels)
```

```python
model = AgglomerativeClustering(n_clusters=None,distance_threshold=2)
```

```python
clust_lab2 = model.fit_predict(scaled_df)
```

```python
pd.DataFrame(clust_lab2).value_counts()
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,10))
dendo = dendrogram(linkage_matrix,truncate_mode='lastp',p=12)
plt.show()
```

```python
plt.figure(figsize=(20,10))
dendo = dendrogram(linkage_matrix,truncate_mode='level',p=5)
plt.show()
```
