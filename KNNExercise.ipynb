```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df2.drop('label2',axis=1), df2['label2'], test_size=0.1, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
scaler = StandardScaler()
knn = KNeighborsClassifier()
```

```python
operations = [('scaler',scaler),('knn',knn)]
```

```python
pipe = Pipeline(operations)
```

```python
k_values = list(range(1,20))
```

```python
param_grid = {'knn__n_neighbors':k_values}
```

```python
full_cv_classifier = GridSearchCV(pipe, param_grid,
                                  cv=5, scoring='accuracy')
```

```python
full_cv_classifier.fit(X_train, y_train)
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
predict_val = full_cv_classifier.predict(X_test)
```

```python
print(classification_report(y_test, predict_val))
```

```python
pd.DataFrame(full_cv_classifier.cv_results_)['mean_test_score'].plot()
```

```python
confusion_matrix(y_test, predict_val)
```
